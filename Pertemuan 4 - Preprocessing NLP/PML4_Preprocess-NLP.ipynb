# PREPROCESSING / PRAPEMROSESAN PADA NLP

Preprocessing merupakan sekumpulan kegiatan yang melibatkan dokumen teks yang bertujuan untuk menghilangkan beberapa bentuk teks khusus, seperti bilangan, tanggal, kata depan, artikel, kata ganti, dan kata umum lainnya. (Kannan & Gurusamy, 2014)

```python
# Import Pandas
import pandas as pd

# Settingan di Pandas untuk menghilangkan warning
pd.options.mode.chained_assignment = None  # default='warn'
```

```python
# Read Data

filename = "DatasetReviewProduk-LabPML.csv"
df = pd.read_csv(filename)
df
```

```python
# Mengambil hanya data teks

data = df[['teks']]
data.head()
```

## Case Folding

Tahap case folding merupakan tahap dimana seluruh huruf pada teks akan digantikan menjadi huruf kecil.

```python
#Case Folding

data['lower'] = data['teks'].str.lower()
data
```

# Punctual Removal

Tahap ini merupakan proses menghilangkan simbol-simbol, tanda baca, dan bagian yang sudah tidak relevan dari teks.

```python
# Punctual Removal

data['punctual'] = data['lower'].str.replace('[^a-zA-Z]+',' ', regex=True)
data
```

# Normalisasi

Tahap normalisasi data diperlukan untuk mengubah format teks dari kata-kata gaul ataupun kata-kata tidak baku menjadi kata-kata standar agar mudah diproses dengan menggunakan kamus bahasa alay yang sudah tersedia. 

```python
# Membaca kamus bahasa alay

# kamus_baku = pd.read_csv('colloquial-indonesian-lexicon.csv')
kamus_baku = pd.read_csv('kata_baku.csv', sep=";")
kamus_baku
```

```python
# Mengubah format dataframe menjadi dictionary

dict_kamus_baku = kamus_baku[['slang','baku']].to_dict('list')
dict_kamus_baku
```

```python
# Memasangkan kata pada dictionary dengan pasangan key-value

dict_kamus_baku = dict(zip(dict_kamus_baku['slang'], dict_kamus_baku['baku']))
dict_kamus_baku
```

```python
# Membuat list baru untuk menampung kalimat2 normal
norm = []

# Dengan perulangan, setiap kalimat akan dipecah menjadi kata2. 
# Kata2 yg terdapat pada dictionary langsung ditukar.
for i in data['punctual']:
    res = " ".join(dict_kamus_baku.get(x, x) for x in str(i).split())
    norm.append(str(res))

# Memasukkan list kalimat normal kembali ke dalam dataframe
data['normalize'] = norm
data
```

# Stopword Removal

Tahap ini merupakan proses pembuangan kata-kata umum yang tidak memiliki makna, misalnya kata depan, konjungsi, kata ganti, kata sambung dan lain-lain.

Proses membuang stopword menggunakan library NLTK (Natural Language Toolkit) yang berbasis bahasa Indonesia.

```python
# Impor Library NLTK

import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')
nltk.download('punkt')
from nltk.tokenize import word_tokenize
```

```python
# Mengambil stopwords Bahasa Indonesia dari NLTK untuk dimasukkan ke himpunan/sets

stop_words = set(stopwords.words('indonesian'))
stop_words
```

```python
# Pengecualian stopwords yang tidak boleh dihapus

hapus = {'tidak','saya','tetapi','meskipun','mirip','dia','belum','kenapa','kalian','semua','ada','kurang','jadi'}
for x in hapus:
    stop_words.remove(x)
```

```python
# Membuat list baru untuk menampung kalimat2 yang sudah dihapus stopword-nya
swr = []

for i in data['normalize']:
    # Setiap kalimat akan dipecah menjadi kata2.
    tokens = word_tokenize(i)
    # Setiap kata akan dimasukkan ke list 'filtered' jika tidak terdapat pada stop_words.
    filtered = [word for word in tokens if word not in stop_words]
    # Kata2 pada list 'filtered' akan digabungkan kembali menjadi kalimat.
    swr.append(" ".join(filtered))

# Memasukkan list kalimat yg bersih kembali ke dalam dataframe
data['stopwords'] = swr
data
```

# Stemming

Tahapan ini bertujuan untuk menghilangkan imbuhan dari kata berimbuhan.

Proses ini menggunakan pustaka (library) Sastrawi. 

```python
# Impor Library
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory

# Mendefenisikan stemmer
factory = StemmerFactory()
stemmer = factory.create_stemmer()

# Membuat list baru untuk menampung kalimat2 yang sudah di-stemming
stem = []

# Dari perulangan, stemmer mengganti setiap kata menjadi stem-nya.
for i in data['stopwords']:
    stem.append(stemmer.stem(str(i)))
    
# Memasukkan list kalimat yg sudah di-stem kembali ke dalam dataframe
data['stemmed'] = stem
data
```

# Tokenisasi

Tahapan ini merupakan proses pemecahan sekelompok tulisan menjadi kata, kelompok kata, ataupun elemen lain yang disebut dengan token.

```python
# Mengecek panjang kalimat maksimum

longest_string = max(data['stemmed'].values.tolist(), key=len)
print(len(longest_string.split()))

print(longest_string)
```

```python
# Mendefinisikan variabel penting

MAX_NB_WORDS = 100000 #Maximum jumlah kata pada vocabulary yang akan dibuat
max_seq_len = 58 #Panjang kalimat maximum
```

```python
# Import Library Tokenizer
from keras.preprocessing.text import Tokenizer
from keras.preprocessing.sequence import pad_sequences
```

```python
# Tokenisasi Training

def tokenize_training(data_training):
    global tokenizer #Menggunakan variabel global agar 'tokenizer' bisa dipake di luar fungsi ini
    tokenizer = Tokenizer(num_words = MAX_NB_WORDS, char_level=False)
    tokenizer.fit_on_texts(data_training['stemmed'])
    word_index = tokenizer.word_index
    
    train_sequences = tokenizer.texts_to_sequences(data_training['stemmed'])
    word_seq_train = pad_sequences(train_sequences, maxlen = max_seq_len)

    return word_index, word_seq_train

word_index, word_seq_train = tokenize_training(data)
data['tokenize'] = word_seq_train.tolist()
data
```

```python
word_index
```

```python
# Tokenisasi Testing

def tokenize_testing(data_testing):
    test_sequences = tokenizer.texts_to_sequences(data_testing['stemmed'])
    word_seq_test = pad_sequences(test_sequences, maxlen = max_seq_len)

    return word_seq_test
```

```python
n = 12
teks = data['teks'][n]
teks_lower = data['lower'][n]
teks_punctual = data['punctual'][n]
teks_normalize = data['normalize'][n]
teks_stopwords = data['stopwords'][n]
teks_stemmed = data['stemmed'][n]
teks_tokenized = str(word_seq_train[n])

print('TEKS :')
print(teks)
print('CASE FOLDING :')
print(teks_lower)
print('PUNCTUAL REMOVAL :')
print(teks_punctual)
print('NORMALIZATION :')
print(teks_normalize)
print('STOPWORD REMOVAL :')
print(teks_stopwords)
print('STEMMING :')
print(teks_stemmed)
print('TOKENIZATION :')
print(teks_tokenized)
```
