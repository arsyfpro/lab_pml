# PERSIAPAN PENGOLAHAN DATA GAMBAR

Data gambar yang dikumpulkan biasanya masih "berantakan" dan berasal dari berbagai sumber yang berbeda.

Agar dapat diproses oleh mesin, data gambar perlu distandarisasi dan diubah ke dalam bentuk angka-angka.

## Import Libraries

Mengimpor semua library yang diperlukan.

```python
# Libraries yang berkaitan dengan files
import zipfile
import os

# Libraries untuk gambar
import numpy as np
import cv2  # pip install opencv-python
import matplotlib.pyplot as plt
```

## Load Dataset

Small RPS Dataset merupakan dataset gambar yang berisi Rock, Paper dan Scissors. Dataset ini merupakan dataset yang sudah diperkecil dari dataset originalnya (created by Julien de la Bru√®re-Terreault, hosted at https://www.kaggle.com/drgfreeman/rockpaperscissors) Dataset ini berisi 897 gambar.

```python
# Mengekstrak file dari ZIP

file = 'rockpaperscissors.zip'
zip_ref = zipfile.ZipFile(file, 'r') # r = read
zip_ref.extractall('')
zip_ref.close()
```

```python
# Melihat isi direktori utama 

main_dir = './' 
# titik slash artinya di direktori yang sama dengan file ini

os.listdir(main_dir)
```

```python
# Melihat isi direktori 'SmallRPSDataset-master'

data_path = './SmallRPSDataset-master' 
os.listdir(data_path)
```

```python
# Mendefinisikan beberapa variabel

kategori = ['paper','rock','scissors'] #list kategori
img_data_list = [] #list kosong untuk menampung data pixel tiap gambar
img_label = [] #list kosong untuk menampung data label tiap gambar dalam bentuk numerik
label_map = {} #dictionary kosong untuk memetakan label numerik ke label string
```

## Pengubahan Gambar menjadi Angka

<ol>
    <li>Masuk ke folder setiap jenis gambar.</li>
    <li>Melakukan mapping label numerik ke label string.</li>
    <li>Membaca setiap gambar ke dalam format angka.</li>
    <li>Meresize setiap gambar.</li>
    <li>Memasukkan gambar hasil resize ke list.</li>
    <li>Memasukkan label ke list.</li>
    <li>Mengubah list gambar dan list label ke dalam format Numpy Array.</li>
    <li>Melakukan normalisasi nilai pixel.</li>
</ol>

```python
# Memberikan label untuk setiap gambarnya

for jenis in kategori:
    # Melihat isi folder setiap jenis: ['paper','rock','scissors']
    img_list = os.listdir(data_path+'/'+ jenis) 
    
    # Membuat dictionary yang berisi label numerik dan label string dari setiap jenis
    i = kategori.index(jenis)
    label_map[i] = str(jenis)

    # Perulangan untuk memasukkan semua file gambar ke dalam img_data_list, dan label numeriknya ke img_label
    for img in img_list:
        input_img = cv2.imread(data_path + '/'+ jenis + '/'+ img )
        input_img_resize = cv2.resize(input_img,(128,128))
        img_data_list.append(input_img_resize)
        img_label.append(i)
```

```python
label_map
```

```python
img_data_list
```

```python
# Mengubah img_data_list ke Numpy Array
img_data = np.array(img_data_list)
# Normalisasi Data = mengubah range dari intensiti pixel
img_data = img_data.astype('float')/255

# Mengubah img_label ke Numpy Array
img_label = np.array(img_label)
```

```python
img_data
```

```python
img_label
```

## Pembagian/Split Dataset

Menggunakan Library train_test_split untuk membagi dataset menjadi 80% dan 20%

```python
from sklearn.model_selection import train_test_split

X = img_data
Y = img_label 

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
print('Ukuran Training Data:', X_train.shape)
print('Ukuran Testing Data:', X_test.shape)
```
