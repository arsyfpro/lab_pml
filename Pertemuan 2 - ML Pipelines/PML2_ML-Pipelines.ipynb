## 4. Feature Engineering (Modifikasi Fitur)

Dataset terdiri dari beraneka ragam fitur (kolom).

Kolom = feature/fitur = atribut<br>
Baris = instance = data

Beberapa teknik yang dapat dilakukan:
<ul>
    <li>Menghapus kolom-kolom yang benar-benar tidak penting dalam proses machine learning, mis: ID, Nama, Nomor Identifikasi, dll.</li>
    <li>Mengubah nama kolom.</li>
    <li>Mengubah data kategorikal menjadi angka.</li>
    <li>dsb.</li>
</ul>


```python
# Menghapus Kolom "ID" dan "Unnamed: 32"

df.drop(['id', 'Unnamed: 32'], inplace=True, axis=1)
df
```

```python
# Encoding Categorical Data (Mengubah data kategorikal menjadi angka)

diagnosis_dict = {'B':0, 'M':1}
df['diagnosis'] = df['diagnosis'].replace(diagnosis_dict)
df
```

## 5A. Pembagian Data (Data Splitting)

Dataset akan dibagi menjadi data training dan data testing.

Umumnya pembagian data training dan data testing adalah 80%:20% atau 90%:10%. <br>
Namun tidak terbatas pada dua bentuk tersebut, <b>yang penting data training harus jauh lebih banyak dari data testing.</b> <br>

Pembagian data tidak boleh 50%:50% atau 1:1 karena <br>
<b>model akan menangkap pola yang terdapat di sebagian dataset (data training) yg merepresentasikan keseluruhan dataset.</b>


```python
from sklearn.model_selection import train_test_split

df_training, df_testing = train_test_split(df, test_size=0.2, random_state=42, shuffle=True)

df_training = df_training.reset_index()
df_testing = df_testing.reset_index()

print('Jumlah data training: ', df_training.shape[0])
print('Jumlah data testing: ', df_testing.shape[0])
```

```python
df_training
```

```python
# Mendapatkan semua kolom kecuali 'diagnosis' pada df_training (sebagai X_train)
X_train = df_training.loc[:, df_training.columns != 'diagnosis']
X_train = df_training.to_numpy()
X_train
```

```python
# Mendapatkan semua kolom kecuali 'diagnosis' pada df_testing (sebagai X_test)
X_test = df_testing.loc[:, df_testing.columns != 'diagnosis']
X_test = df_testing.to_numpy()
X_test
```

```python
# Mendapatkan kolom 'diagnosis' pada df_training (sebagai Y_train)
Y_train = df_training['diagnosis'].to_numpy()
Y_train
```

```python
# Mendapatkan kolom 'diagnosis' pada df_testing (sebagai Y_test)
Y_test = df_testing['diagnosis'].to_numpy()
Y_test
```

## 5B. Training Model

Memilih algoritma yang sesuai dengan dataset.

Pemilihan algoritma harus berdasarkan kelebihan dari algoritma tersebut (<b>Harus ada alasan tertulis dari karya ilmiah orang lain</b>).

Akurasi tinggi dari suatu algoritma yang tertera pada karya ilmiah orang lain <b>tidak bisa dijadikan satu-satunya alasan kamu memilih algoritma tersebut</b>, melainkan hanya sebagai informasi pendamping saja.


```python
# Algoritma K-Nearest Neighbor (KNN)

from sklearn.neighbors import KNeighborsClassifier
knn_model = KNeighborsClassifier(n_neighbors=5)
knn_model.fit(X_train, Y_train)
```

```python
# Algoritma Support Vector Machine (SVM)

from sklearn.svm import SVC
svm_model = SVC(gamma = 'auto', kernel="linear", C=2)
svm_model.fit(X_train, Y_train)
```

```python
# Algoritma Logistic Regression

from sklearn.linear_model import LogisticRegression
logr_model = LogisticRegression(random_state=0)
logr_model.fit(X_train, Y_train)
```

```python
# Algoritma Decision Tree

from sklearn.tree import DecisionTreeClassifier
dtr_model = DecisionTreeClassifier(criterion="entropy", random_state=0)
dtr_model.fit(X_train,Y_train)
```

```python
# Algoritma Random Forest

from sklearn.ensemble import RandomForestClassifier
rf_model = RandomForestClassifier(n_estimators = 300, random_state = 42)
rf_model.fit(X_train, Y_train)
```

```python
# Algoritma XGBoost

import xgboost as xgb
xgb_model = xgb.XGBClassifier(objective="binary:logistic", random_state=42, max_depth=10, learning_rate=1e-3)
xgb_model.fit(X_train, Y_train)
```

## 6. Evaluasi Model (Testing)

Tahap ini dilakukan dengan cara melakukan prediksi terhadap data testing, lalu dibandingkan dengan label sebenarnya.

Prediksi dilakukan karena kita hanya menggunakan nilai X_test tanpa mengetahui/dipengaruhi oleh Y_test.<br>
Y_test hanya akan dipakai untuk membandingkan hasil.

```python
# Memilih model yang digunakan: knn_model, svm_model, logr_model, dtr_model, rf_model, xgb_model
model = knn_model

# Melakukan prediksi terhadap Data Testing
predictions = model.predict(X_test)
predictions
```

```python
# Memasukkan hasil prediksi (predictions) ke dalam tabel df_testing
df_hasil = df_testing
df_hasil['diagnosis_pred'] = predictions
df_hasil[['diagnosis', 'diagnosis_pred']]
```

```python
#  Menampilkan Confusion Matrix dalam format matriks

from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_true=df_hasil['diagnosis'], y_pred=df_hasil['diagnosis_pred'])
print(cm)
```

```python
#  Menampilkan Confusion Matrix dalam grafik heatmap

import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8,6))
ax= plt.subplot()
sns.heatmap(cm, annot=True, fmt='g', ax=ax) #annot=True to annotate cells, ftm='g' to disable scientific notation

# labels, title and ticks
ax.set_xlabel('Predicted labels')
ax.set_ylabel('True labels')
ax.set_title('Confusion Matrix')
ax.xaxis.set_ticklabels(['Benign (0)', 'Malignant (1)'])
ax.yaxis.set_ticklabels(['Benign (0)', 'Malignant (1)'])

plt.show()
```

```python
#  Menghitung nilai akurasi

import sklearn
print("Accuracy :")
print(sklearn.metrics.accuracy_score(df_hasil['diagnosis'], df_hasil['diagnosis_pred']))
```
