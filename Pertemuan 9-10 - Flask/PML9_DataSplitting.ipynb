Langkah awal yang harus dilakukan adalah mempersiapkan data latih dan data uji (di luar sistem).

```python
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
filename = "../Pertemuan 4 - Preprocessing NLP/DatasetReviewProduk-LabPML.csv"

data = pd.read_csv(filename)
data
```

```python
data_training, data_testing = train_test_split(data, test_size=0.2, random_state=42, shuffle=True)
```

```python
data_training.to_csv('datax_training.csv', index=False)
data_testing.to_csv('datax_testing.csv', index=False)
```
